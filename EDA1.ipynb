## Outliers treatment

```python
#decide whether to do outlier treatment or not
```

```python
def wisker(col):
    q1,q3 = np.percentile(col,[25,75])
    iqr = q3-q1
    lw = q1-1.5*iqr
    uw = q3+1.5*iqr
    return lw,uw
```

```python
wisker(df['AC'])
```

```python
for i in ['AC','UC','DL', 'ALTV']:
    lw,uw = wisker(df[i])
    df[i] = np.where(df[i]<lw,lw,df[i])
    df[i] = np.where(df[i]>lw,uw,df[i])
```

```python
for i in ['AC','UC','DL', 'ALTV']:
    sns.boxplot(df[i])
    plt.show()
```
